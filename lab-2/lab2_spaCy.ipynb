Task1: This section is to load transcript from lab1 of github.

```python
import requests

# Raw GitHub link to transcript.txt
url = "https://raw.githubusercontent.com/nabinjoshi54/lis5693/main/lab-1/transcript.txt"

# Send request to GitHub
response = requests.get(url)

# Raise error if link fails
response.raise_for_status()

# Store file content as text
text = response.text

print("Transcript successfully loaded!")
```

Task2: This is to count total characters and pring firsst 100 characters.

```python
# Count total number of characters
print("Total number of characters:", len(text))

# Print first 100 characters
print("\nFirst 100 characters of the transcript:\n")
print(text[:100])
```

Task3: This section perform sentence segmentation using a blank spaCy pipeline

```python
import spacy

# Create blank English pipeline
nlp_blank = spacy.blank("en")

# Add sentencizer for rule-based sentence segmentation
nlp_blank.add_pipe("sentencizer")

# Process text
doc_blank = nlp_blank(text)

# Convert sentences to list
sentences = list(doc_blank.sents)

# Print results
print("Total number of sentences:", len(sentences))

print("\nFirst 3 sentences:\n")
for sent in sentences[:3]:
    print("-", sent.text.strip())
```

Task4: This section perform word count and token analysis.

```python
# Extract alphabetic tokens only
words = [token.text.lower() for token in doc_blank if token.is_alpha]

# Count total words
total_words = len(words)

# Count unique words
unique_words = len(set(words))

print("Total number of words:", total_words)
print("Number of unique words:", unique_words)
```

Task5: The code in this block find the most frequent word.

```python
from collections import Counter

# Count word frequencies
word_freq = Counter(words)

# Most frequent word
most_common_word, most_common_count = word_freq.most_common(1)[0]

print("Most frequent word:", most_common_word)
print("Frequency:", most_common_count)

print("\nTop 10 most frequent words:\n")
for word, count in word_freq.most_common(10):
    print(word, "-", count)
```

Task6: This will run the full spaCy pipeline and extract named entities.

```python
import spacy
from collections import Counter

# Load full spaCy model
nlp_full = spacy.load("en_core_web_sm")

# Process text
doc_full = nlp_full(text)

# Count named entities
print("Total named entities found:", len(doc_full.ents))

# Count entity types
entity_labels = [ent.label_ for ent in doc_full.ents]
label_counts = Counter(entity_labels)

print("\nEntity types and their counts:\n")
for label, count in label_counts.most_common():
    print(label, "-", count)

print("\nFirst 10 named entities:\n")
for ent in doc_full.ents[:10]:
    print(ent.text, "-", ent.label_)
```

Task7: This will use phrasematcher from spacy.matcher to find topic related phrases.

```python
from spacy.matcher import PhraseMatcher

# Replace these with your topic-specific phrases
phrases = [
    "data science",
    "communication",
    "conversation"
]

matcher = PhraseMatcher(nlp_full.vocab, attr="LOWER")

patterns = [nlp_full.make_doc(p) for p in phrases]
matcher.add("TOPIC_PHRASES", patterns)

matches = matcher(doc_full)

print("Total matches found:", len(matches))

for match_id, start, end in matches[:15]:  # limit output
    span = doc_full[start:end]
    print("\nMatch:", span.text)
    print("Sentence:", span.sent.text.strip())
```
