Setup: meaning installing libraries. I am using pandas to load and manipulate the Seattle Library dataset. Then, I will use ALtair to create interactive visualization. I will load dataset into a dataframe named seattle_df.
I will use the same "Seattle Library Data" to visualize at least two different columns. The link to the data set is: "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"
This is what it looks like when I open the link to dataset.

```python
import pandas as pd
import altair as alt
import requests, io

# Allow larger datasets in Altair (avoids max row limits)
alt.data_transformers.disable_max_rows()
```

```python
url = "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"
response = requests.get(url)
response.raise_for_status()

seattle_df = pd.read_csv(io.StringIO(response.text))
seattle_df.head()
```

Task-1: For this task, my job is to visualize at least two different columns from seattle_df. Since the professor already used 'Material Type', I will use 'Checckoutyear', 'UsageClass', and 'Checkouts'. First, I am grouping the dataset using pandas to aggregate the total number of checkouts for each year and usage category. This will help data structured format that can be used for visualization.  

```python
year_usage = (
    seattle_df.groupby(["CheckoutYear","UsageClass"], as_index=False)["Checkouts"]
    .sum()
)

year_usage.head()
```

Task2: I am using differnet Altair Marks.
For visualization 1, I will do line chard and for this, 'CheckoutYear' and 'Checkpouts' are being used. It will show the trend of total library checkouts over time by year.
For visualization 2, I will do area chart and for this, all three collumns are being used -- CheckoutYear, UsageClass, and Checkouts. This visualization compare the number of checkouts between different usage classes.

```python
year_totals = (
    seattle_df.groupby("CheckoutYear", as_index=False)["Checkouts"]
    .sum()
)

chart1 = alt.Chart(year_totals).mark_line(point=True).encode(
    x='CheckoutYear:O',
    y='Checkouts:Q',
    tooltip=['CheckoutYear','Checkouts']
).properties(
    title="Total Library Checkouts per Year",
    width=400,
    height=300
)

chart1
```

```python
chart2 = alt.Chart(year_usage).mark_area(opacity=0.6).encode(
    x='CheckoutYear:O',
    y='Checkouts:Q',
    color='UsageClass:N',
    tooltip=['CheckoutYear','UsageClass','Checkouts']
).properties(
    title="Physical vs Digital Checkouts Over Time",
    width=400,
    height=300
)

chart2
```

Task-3: I am adding interactivity to the visualizations to make them more informative and easier to explore. Tooltips will be added so that exact values appear when hovering over the chart. I am using Altair's .interactive() function. For the area chart, I will add a legend selection so users can click a category in the legend to highlight the data.

```python
chart1 = chart1.interactive()

selection = alt.selection_point(fields=['UsageClass'], bind='legend')

chart2 = alt.Chart(year_usage).mark_area(opacity=0.6).encode(
    x='CheckoutYear:O',
    y='Checkouts:Q',
    color='UsageClass:N',
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2)),
    tooltip=['CheckoutYear','UsageClass','Checkouts']
).add_params(selection)

chart2
```

Task-4: In this task, I will combine the two visualizations in a single view or frame using Altair's horizontal concatenation feature. The task help compare both trends easily. The charts will be combined using | operator, which places one chart next to the other in the final visualization.

```python
chart1 | chart2
```
